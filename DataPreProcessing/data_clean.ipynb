```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Data/Forest_Fire_2019_20.csv"

# Read Purchasing File and store into Pandas data frame
fire_data_df = pd.read_csv(file_to_load)

fire_data_df
```

```python
fire_data_drop_df = fire_data_df.dropna()
fire_data_drop_df
```

```python
fire_clean_df = pd.DataFrame({"State":fire_data_drop_df["Return to Index"],
    "TotalForestHectares":fire_data_drop_df["Unnamed: 4"],
    "TotalForestBurntHectares":fire_data_drop_df["Unnamed: 8"]})                    
fire_clean_df
```

```python
print (fire_clean_df.dtypes)
```

```python
fire_clean_df['TotalForestBurntHectares'].astype(str)
fire_clean_df['TotalForestHectares'].astype(str)
repl_dict={",":""}
fire_clean_df["TotalForestBurntHectares"]=fire_clean_df["TotalForestBurntHectares"].replace(repl_dict,regex=True)
fire_clean_df["TotalForestHectares"]=fire_clean_df["TotalForestHectares"].replace(repl_dict,regex=True)
fire_clean_df
```

```python
fire_clean_df['TotalForestBurntHectares'] =fire_clean_df['TotalForestBurntHectares'].astype(float)
fire_clean_df['TotalForestHectares'] = fire_clean_df['TotalForestHectares'].astype(float)
fire_clean_df
```

```python
print(fire_clean_df.dtypes)
```

```python
fire_clean_df['PercentForestBurnt'] = fire_clean_df['TotalForestBurntHectares']/fire_clean_df['TotalForestHectares']*100.
fire_clean_df['PercentForestBurnt'] = fire_clean_df['PercentForestBurnt'].round(2)
fire_clean_df
```

```python
fire_clean_df.reset_index(drop=True)
```

```python
fire_clean_df.to_csv("clean_fire_data.csv")
```
